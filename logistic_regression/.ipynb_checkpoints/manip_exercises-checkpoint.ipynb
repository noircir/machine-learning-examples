```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# ex. 1

A = np.array([
  [0.3, 0.6, 0.1],
  [0.5, 0.2, 0.3],
  [0.4, 0.1, 0.5]])

v = np.ones(3) / 3

num_iters = 25
distances = np.zeros(num_iters)
for i in range(num_iters):
    v2 = v.dot(A)
    d = np.linalg.norm(v2 - v)
    distances[i] = d
    v = v2

plt.plot(distances)
```

```python
# ex.2

def sampleY(n=1000):
    # draw n samples from uniform dist.
    X = np.random.random(n)
    Y = X.sum()
    return Y


# now draw N Y's
N = 1000
Y_samples = np.zeros(N)
for i in range(N):
    Y_samples[i] = sampleY()


# now plot the Y_samples
plt.hist(Y_samples, bins=20)
plt.show()
```

```python
# ex. 6 XOR

N = 2000
X = np.random.random((N, 2))*2 -1

# generate labels
Y = np.zeros(N)
Y[(X[:,0] < 0) & (X[:,1] > 0)] = 1
Y[(X[:,0] > 0) & (X[:,1] < 0)] = 1

# plot it
plt.scatter(X[:,0], X[:,1], c=Y)
```
