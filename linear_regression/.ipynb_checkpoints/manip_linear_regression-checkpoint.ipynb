```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the data
X = []
Y = []

for line in open('data_1d.csv'):
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
# turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# plot to see what they look like
plt.scatter(X,Y)
```

```python
# apply the equations we learned to calculate a and b
denom = X.dot(X) - X.mean() * X.sum()
a = ( X.dot(Y) - Y.mean() * X.sum() ) / denom
b = ( Y.mean() * X.dot(X) - X.mean() * X.dot(Y) ) / denom
```

```python
print(a)
print(b)
```

```python
# calculate predicted Y
Y_hat = a*X + b
```

```python
# plot it all
plt.scatter(X,Y,color='blue')
plt.plot(X,Y_hat,color='r')
```

```python
# calculate r-squared
d1 = Y - Y_hat
d2 = Y - Y.mean()

r2 = 1 - d1.dot(d1) / d2.dot(d2)
r2  # very close to 1, the perfect model.
```

## Moore's law

```python
# Moore's law states that processor speeds double every two years
import re

X = []
Y = []

non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t')
    
    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
    
X = np.array(X)
Y = np.array(Y)

# plot of raw data
plt.scatter(X, Y)
```

Some nonlinear regression problems can be moved to a linear domain by a suitable transformation of the model formulation. If we take the logarithm of both sides, it becomes a linear model.

https://en.wikipedia.org/wiki/Nonlinear_regression#Linearization

If a variable x exhibits exponential growth, then the log (to any base) of x grows linearly over time, as can be seen by taking logarithms of both sides of the exponential growth equation:

https://en.wikipedia.org/wiki/Exponential_growth

```python
Y = np.log(Y)
plt.scatter(X, Y)
```

```python
denom = X.dot(X) - X.mean() * X.sum()
a = ( X.dot(Y) - Y.mean() * X.sum() ) / denom
b = ( Y.mean() * X.dot(X) - X.mean() * X.dot(Y) ) / denom
```

```python
Y_hat = a*X + b
plt.scatter(X,Y,color='blue')
plt.plot(X,Y_hat,color='r')
```

```python
# r-square metrics
d1 = Y - Y_hat
d2 = Y - Y.mean()

r2 = 1 - d1.dot(d1) / d2.dot(d2)
r2 
```
